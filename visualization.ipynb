```python
import matplotlib.pyplot as plt

# Sample dataset
labels = ['Apple', 'Banana', 'Orange', 'Strawberry']
sizes = [35, 25, 20, 20]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Add a title
plt.title('Fruit Consumption')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset
data = np.random.rand(10, 10)
print(data)
# Create the heat map
plt.imshow(data, cmap='YlGnBu')

# Add a colorbar
plt.colorbar()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample dataset
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add a title and labels for the axes
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample dataset
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y)

# Add a title and labels for the axes
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```
