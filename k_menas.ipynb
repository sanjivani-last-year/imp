```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv('1st.csv')
df
```

```python
df.isnull()
```

```python
df
```

```python
plt.scatter(df.C1,df.C2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df[['C1']])
df['C1'] = scaler.transform(df[['C1']])

scaler.fit(df[['C2']])
df['C2'] = scaler.transform(df[['C2']])

df.head()
```

```python
plt.scatter(df.C1,df.C2)
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['C1','C2']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df1
```

```python
plt.scatter(df1.C1,df1.C2,c="green",s=100)
plt.scatter(df2.C1,df2.C2,c="red",s=100)
plt.scatter(df3.C1,df3.C2,c="black",s=100)
plt.scatter(km.cluster_centers_[:0],km.cluster_centers_[:1],color='purple',marker='*',label='centroid')
plt.xlabel('C1')
plt.ylabel('C2')
plt.legend()
```
